```python
import fnmatch, os, shutil

root_path = r"D:\research work\Data\ADNIorigCOPY"
new_path = r"D:\research work\Data\required_ADNI_data"

#file_name = "strip.img"
file_name = "strip.hdr"
base_file_pattern = "sc-1"

total_files = 0
for root, dirs, files in os.walk(root_path):
    if (base_file_pattern in root):
        for filename in fnmatch.filter(files, file_name):
            src = os.path.join(root, filename)
            dest = src.replace(root_path,"")
            dest = dest.replace("\\","_")
            dest = dest[1:]
            dest = os.path.join(new_path, dest)
            print(src)
            print(dest)
            shutil.copy2(src, dest)
            total_files = total_files + 1
            print("+++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++")
            
print(total_files)
```
