```python
import fnmatch,os, gzip, shutil
rootPath = r"D:\research work\data\ADNIorigCOPY"
pattern = 'strip*.gz'
        
def gunzip(file_path, output_path):
    with gzip.open(file_path, "rb") as f_in, open(output_path, "wb") as f_out:
        shutil.copyfileobj(f_in, f_out)

totalStripFileCount = 0
baseStripFileCount = 0
for root,dirs,files in os.walk(rootPath):
    for filename in fnmatch.filter(files, pattern):
        totalStripFileCount = totalStripFileCount+1
        if (root.find("sc-1") != -1):
            baseStripFileCount = baseStripFileCount+1
        print(os.path.join(root, filename))
        tempFileName = os.path.join(root, filename)
        gunzip(tempFileName, tempFileName.replace(".gz",""))

print(totalStripFileCount)
print(baseStripFileCount)
       
```
